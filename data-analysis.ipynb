```python
plt.style.use('fivethirtyeight')
data=tita.Survived.value_counts()
plt.figure(figsize=(10,7))
plt.title("survived vs dead",fontsize=20)
fig=data.plot.bar(color=['lightblue','green'])
plt.xlabel('survived',fontsize=15)
plt.ylabel('count',fontsize=15)
plt.ylim(0,650)
plt.xticks([0,1],['dead','survived'],Rotation=45,fontsize=15)
plt.yticks(np.arange(0,650,50))
plt.annotate(data[0],xy=(-0.05,data[0]+5),fontsize=14)
plt.annotate(data[1],xy=(0.95,data[1]+5),fontsize=14)
plt.show()
```

computing the correlation matrix

```python
import pandas as pd
df = pd.read_excel(r"C:\Users\Naveen\Desktop\data sets\cricket highest scores.xlsx")
df.head()
```

```python
corr=df.corr()
corr
```

```python
sns.heatmap(corr)
```

```python
import numpy as np
import seaborn as sns
mask=np.zeros_like(corr)
mask[np.triu_indices_from('mask')]=True
with sns.axes_style('white'):
    f,ax=plt.subplots(7,5)
    ax=sns.heatmap(corr,mask=mask,vmax=1,square=True)
```
